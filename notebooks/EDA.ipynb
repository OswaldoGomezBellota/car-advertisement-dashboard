```python
import pandas as pd # Librería para manipulación de datos
import plotly.express as px # Librería para gráficos interactivos (Dispersión, Histograma, Barra, Linea, etc.)
import plotly.io as pio # Configurar Plotly para mostrar gráficos en el navegador
import streamlit as st # Para construir el dashboard

# Cargar el dataset
car_data = pd.read_csv('../vehicles_us.csv')

# Histograma del kilometraje
hist_fig = px.histogram(car_data, x="odometer",
                        title="Distribución del kilometraje de los vehículos")
hist_fig

# Relación precio vs kilometraje
scatter_fig = px.scatter(car_data, x="odometer", y="price",
                         title="Relación entre precio y kilometraje")
scatter_fig
```
